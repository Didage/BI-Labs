```python
df_tracks_t = df_tracks

# Ahora definimos la función que nos va a permitir construir nuestra clase.
def smoking_label (row):
    if row['Smoking'] == 'No' or  'No' in str(row['Smoking']):
        return 0
    return 1
df_tracks_t['Smoking']=df_tracks_t.apply (lambda row: smoking_label(row), axis=1) 
```

```python
# Ahora definimos la función que nos va a permitir construir nuestra clase.
def transform_label (row,label):
    if row[label] == 'No' or  'No' in str(row[label]):
        return 0
    return 1 
```

```python
df_tracks_t['AlcoholDrinking']=df_tracks_t.apply (lambda row: transform_label(row,'AlcoholDrinking'), axis=1) 
```

```python
df_tracks_t['Stroke']=df_tracks_t.apply (lambda row: transform_label(row,'Stroke'), axis=1) 
df_tracks_t['DiffWalking']=df_tracks_t.apply (lambda row: transform_label(row,'DiffWalking'), axis=1) 
df_tracks_t['Diabetic']=df_tracks_t.apply (lambda row: transform_label(row,'Diabetic'), axis=1)
df_tracks_t['PhysicalActivity']=df_tracks_t.apply (lambda row: transform_label(row,'PhysicalActivity'), axis=1) 
df_tracks_t['Asthma']=df_tracks_t.apply (lambda row: transform_label(row,'Asthma'), axis=1) 
df_tracks_t['KidneyDisease']=df_tracks_t.apply (lambda row: transform_label(row,'KidneyDisease'), axis=1) 
df_tracks_t['SkinCancer']=df_tracks_t.apply (lambda row: transform_label(row,'SkinCancer'), axis=1) 
```

```python
display(df_tracks_t.sample(5)) # Muestra
```

```python
def transform_age (row):
    try:
        edad = int((str(row['AgeCategory']).split('-'))[0])
    except:
        return 70
    return edad
```

```python
df_tracks_t['AgeCategory']=df_tracks_t.apply (lambda row: transform_age(row), axis=1) 
```

```python
df_tracks_t['HeartDisease']=df_tracks_t.apply (lambda row: transform_label(row,'HeartDisease'), axis=1) 
```
